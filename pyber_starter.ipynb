```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

dataframe2 = pd.merge(ride_data,city_data, on = 'city', how = 'left')

dataframe2.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_type = dataframe2[dataframe2["type"] == "Urban"].groupby([dataframe2["city"]])
rural_type = dataframe2[dataframe2["type"] == "Rural"].groupby([dataframe2["city"]])
suburban_type = dataframe2[dataframe2["type"] == "Suburban"].groupby([dataframe2["city"]])

urban_x = urban_type["ride_id"].count()
urban_y = urban_type["fare"].mean()
urban_s = urban_type["driver_count"].mean()

rural_x = rural_type["ride_id"].count()
rural_y = rural_type["fare"].mean()
rural_s = rural_type["driver_count"].mean()

suburban_x = suburban_type["ride_id"].count()
suburban_y = suburban_type["fare"].mean()
suburban_s = suburban_type["driver_count"].mean()



# Build the scatter plots for each city types

plt.scatter(urban_x, urban_y, label = "Urban", s=urban_s * 8, color = "blue", edgecolor = "black", alpha = 1.00, marker = 'o')
plt.scatter(rural_x,rural_y, label = "Rural", s=rural_s * 8, color = "red", edgecolor = "black",alpha = 1.00, marker = 'o')
plt.scatter(suburban_x, suburban_y, label = "Suburban", s=suburban_s * 8, color = "green", edgecolor = "black", alpha = 1.00, marker = 'o')

plt.grid()

# Incorporate the other graph properties
plt.xlabel  = ("Rides per city")
plt.ylabel  = ("Average fare")
plt.title ("Ride data")

# Create a legend
legend = plt.legend(fontsize = 10, loc = "best", title = "City Types")

# Incorporate a text label regarding circle size
#plt.txt(42,35,"Circle size corresponds with driver count", fontsize = 8)

# Save Figure
plt.savefig

plt.show
```
